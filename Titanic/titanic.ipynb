```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
print('Number of rows in train data:', len(train_df))
train_df.head()
```

```python
train_df = train_df.drop(columns = ['PassengerId', 'Ticket', 'Cabin'])
```

```python
train_df.head()
```

```python
names = np.asarray(train_df.Name.values)
title = []
for name in names:
    name = name.split(',')[1]
    name = name.split('.')[0]
    title.append(name)
train_df['Title'] = title
train_df = train_df.drop(columns=['Name'])
train_df.head()
```

```python
for column in ['Age', 'Fare']:
    plt.figure()
    sns.displot(train_df[['Survived', column]], x=column, col = 'Survived')
    plt.show()
```

```python
all_ages = np.asarray(train_df.Age.values)
age_range = []
for age in all_ages:
    if age <= 5:
        age_range.append('baby')
    elif age <= 13:
        age_range.append('child')
    elif age <= 25:
        age_range.append('young adult')
    elif age <= 40:
        age_range.append('adult')
    elif age <= 65:
        age_range.append('middle aged')
    elif age <= 100:
        age_range.append('senior')
    else:
        age_range.append('adult')

train_df['AgeRange'] = age_range
train_df.head()
```

```python
sibs = np.asarray(train_df.SibSp.values)
par = np.asarray(train_df.Parch.values)

family = []
for i in range(len(sibs)):
    total = sibs[i] + par[i]
    family.append(total)
    
train_df['Family'] = family
train_df.head()
```

```python
titles = list(train_df.Title.values)
titles = [x.lstrip() for x in titles]

for i in range(len(titles)):
    if titles[i] in ['Lady', 'the Countess','Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']:
        titles[i] = 'Rare'
    elif titles[i] in ['Mlle', 'Ms']:
        titles[i] = 'Miss'
    elif titles[i] == 'Mme':
        titles[i] = 'Mrs'
        

train_df['Title'] = titles
train_df.head()
```

```python
train_df[['Title', 'Sex']].value_counts().sort_values(ascending=True)
```

```python
train_df[['Title', 'Survived']].groupby('Title', as_index=False).mean().sort_values(
    'Survived', ascending=False)
```

```python
title_mapping = {'Mr': 1, 'Rare':2, 'Master':3, 'Miss':4, 'Mrs': 5}
train_df['Title'] = train_df['Title'].map(title_mapping)
train_df.head()
```

```python
train_df['Title'].fillna(0)
train_df['Sex'] = train_df['Sex'].map({'Male': 0, 'Female':1})
train_df.head()
```
